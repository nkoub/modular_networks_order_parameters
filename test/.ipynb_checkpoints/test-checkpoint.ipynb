# A random modular network

```python
# This graph is not modular, but is used here for illustration purposes
graph_example = nx.erdos_renyi_graph(30, 0.17, seed=2121)
membership_example = np.concatenate((np.zeros(8), np.ones(15), 2*np.ones(7)))

plt.title("Network")
plt.axis("off")
gpos = nx.circular_layout(graph_example)
nx.draw_networkx_nodes(graph_example, pos=gpos, node_size=200, node_color=membership_example)
nx.draw_networkx_edges(graph_example, pos=gpos, edge_color="gray")
```

```python
ax1, ax2 = plt.subplot(121), plt.subplot(122)
plt.subplots_adjust(wspace=0.5)

adj = nx.adjacency_matrix(graph_example).todense()
ax1.set_title("adjacency matrix")
ax1.imshow(adj, 
           cmap=plt.cm.binary, interpolation="nearest",
           aspect="equal", origin="upper")

lapl = -nx.laplacian_matrix(graph_example).todense() # L = A - D
ax2.set_title("Laplacian matrix")
ax2.imshow(lapl, 
           cmap=plt.cm.magma, interpolation="nearest",
           aspect="equal", origin="upper")
```

# Random network of coupled oscillators with Gaussian noise

\begin{equation} 
% \dot{\theta}_i = \omega_i + \frac{K}{N}\sum_{j=1}^{N}\!A_{ij}  \sin(\theta_j - \theta_i)\, + \sqrt{2D}\xi_i(t) 
\end{equation}

\begin{equation} 
\dot{\mathbf{x}}_i = F(\mathbf{x}_i) + \sigma \sum_{j=1}^{N}\!A_{ij}  H(\mathbf{x}_j)\, + \sqrt{2D}\xi_i(t) 
\end{equation}

+ $\mathbf{x}_i$ denotes the state of oscillator $i$ (in the node $i$).
+ $F(\mathbf{x}_i)$ describes the evolution of oscillator $i$.
+ $A_{ij}$ describes how the oscillators are coupled together.
+ $H$ is a function tht describes the coupling.
+ Gaussian white noise: $\langle \xi(t) \rangle = 0$ and $\langle \xi(t)\xi(t-\tau) \rangle = 1$.
 
### Numerical integrations was done with Heun method in C++ (see library https://github.com/nkoub/sodeint/)

# Here we consider a network with 100 nodes and 3 communities


```python
membership = np.loadtxt("membership.txt", dtype=np.int)
print "network size:" , len(membership), "\n"
print "community index of the nodes:\n", membership
```

## We need to **group** and **sort** the nodes in the communities

```python
com_id = np.unique(membership) 

com = {}
for i in com_id:
    com[i] = np.argwhere(membership==i).flatten()

for i in com_id:
    print "Community", i+1, "has", len(com[i]), "nodes:\n", com[i], "\n"
```

## order the nodes by communities

```python
nodes_id = np.concatenate((com[0], com[1], com[2]))
```

```python
spacetime = np.loadtxt("spacetime_test.txt")[nodes_id] # Load the results from the simulations
tsnap = 142
snapshot = spacetime[:, tsnap]

print "Snapshot at t=200" 
plt.scatter(nodes_id, snapshot, c=membership, cmap=plt.cm.viridis)
plt.xlabel("nodes i")
plt.ylabel("phase")
plt.show()
```

```python
print "space:", spacetime.shape[0], "nodes and time:",spacetime.shape[1], "steps"
plt.clf()
cb = plt.imshow(spacetime[nodes_id], cmap=plt.cm.viridis,
                interpolation="nearest", 
                aspect="auto", origin="lower")
cbar = plt.colorbar(cb)
cbar.ax.set_ylabel("phase")
plt.xlabel("time")
plt.ylabel("nodes i")
plt.show()
```

```python
# remove some transient time
spacetime = spacetime[:, 85:]

print "space:", spacetime.shape[0], "nodes and time:", spacetime.shape[1], "steps"
plt.clf()
cb = plt.imshow(spacetime, cmap=plt.cm.viridis,
                interpolation="nearest", 
                aspect="auto", origin="lower")
cbar = plt.colorbar(cb)
cbar.ax.set_ylabel("phase")
plt.xlabel("time")
plt.ylabel("nodes i")
plt.show()
```

# Global order parameters

\begin{equation}
\rho = \frac{1}{N}\sum_{j=1}^N e^{\mathbf{i}\theta_j(t)} 
\end{equation}


```python
op1 = global_order_parameter_spacetime(spacetime[com[0]]).mean()
op2 = global_order_parameter_spacetime(spacetime[com[1]]).mean()
op3 = global_order_parameter_spacetime(spacetime[com[2]]).mean()
op = global_order_parameter_spacetime(spacetime).mean()

print "order parameter  of community 1: {} \
                 \n\t\t of community 2: {} \n\t\t of community 3: {} \
                 \n\t\t of the whole network: {} ".format(op1, op2, op3, op)
```

# Local order parameter for one time snapshot

\begin{equation} 
\rho_i = \frac{1}{2\delta}\sum_{j=i-\delta}^{i+\delta} e^{\mathbf{i}\theta_j(t)} 
\end{equation}

# Mean phase velocity

\begin{equation}
\Omega_i = \frac{2 \pi K_i}{\Delta T}\,,
\end{equation}
where $K_i$ is the number of periods of the $i$-th oscillator during a time interval $\Delta T$.  


```python
omega = mean_phase_velocity_spacetime(spacetime, order=5)

plt.clf()
plt.scatter(nodes_id, omega, c=membership, cmap=plt.cm.viridis)
plt.xlabel("nodes i")
plt.ylabel("mean phase velocity")
plt.show()
```

# Local curvature index

\begin{equation} 
\hat{D}\theta_i(t) := \sum_{j=i-\delta}^{j=i+\delta} \!\left[\theta_j(t) - \theta_i(t)\right]\, 
\end{equation}


The norm of the operator $\hat{D}$ is the local curvature index and:
- $ \|\hat{D}\{\theta_{i_\text{coh}}(t)\}\|\approx 0 $ in the coherent clusters  
- $\|\hat{D}\{\theta_{i_\text{incoh}}(t)\}\|\neq 0$ and has pronounced fluctuations  in the incoherent clusters. 


```python
local_curv = local_curvature_space(snapshot, order=5)

plt.clf()
plt.scatter(nodes_id, local_curv, c=membership, cmap=plt.cm.viridis)
plt.xlabel("nodes i")
plt.ylabel("local curvature")
plt.show()   
```

```python
lcsp = local_curvature_spacetime(spacetime, order=5)

plt.clf()
cb = plt.imshow(lcsp, aspect='auto', origin="lower", cmap=plt.cm.viridis)
cbar = plt.colorbar(cb)
cbar.ax.set_ylabel("local curvature")
plt.xlabel("time")
plt.ylabel("nodes i")
plt.show()
```

# Coherence in time and in space

+ Metastabilite index $\lambda$:

\begin{equation} 
\lambda = {\langle \sigma_{\textbf{met}} \rangle}_{C_m}
\end{equation}
where
\begin{equation} 
\sigma_{\textbf{met}}(m)=\frac{1}{T-1}\sum_{t=1}^T(\rho_m(t)-{\langle \rho_m \rangle}_T)^2
\end{equation}

+ Chimera-like index $\chi$
\begin{equation} 
\chi={\langle \sigma_{\textbf{chi}} \rangle}_T, 
\end{equation}
where
\begin{equation} 
\sigma_{\textbf{chi}}(t)=\frac{1}{M-1}\sum_{m=1}^M(\rho_m(t)-{\langle \rho(t) \rangle}_M)^2 
\end{equation}



```python
# Chimera-like index
chi = chimera_index(spacetime, membership)
print "chimera-like index:",chi
```

```python
# Metastability index
lam = metastability_index(spacetime, membership)
print "metastability index:" ,lam
```

```python
if chi > lam: print "chimera-like states"
if chi < lam: print "metastable states"
```
