```python
import numpy as np
import time
import matplotlib.pyplot as plt
%matplotlib inline

from nktools import fmod_pi
from order_parameters import *
```

```python
spacetime = fmod_pi(np.loadtxt("spacetime.txt")).T
print "space:", spacetime.shape[0], "nodes \ntime:", spacetime.shape[1], "steps"
```

```python
plt.imshow(spacetime,cmap=plt.cm.viridis,
           interpolation="nearest",
           aspect="auto",origin="lower")
plt.xlabel("time")
plt.ylabel("phase")
plt.show()
```

```python
# Global order parameter (slow method)
t0 = time.time()
op1 = global_order_parameter_spacetime_slow(spacetime[:300,-300:]).mean()
op2 = global_order_parameter_spacetime_slow(spacetime[300:,-300:]).mean()
op = global_order_parameter_spacetime_slow(spacetime[:,-300:]).mean()
t1 = time.time()
print "order parameter of community 1:{}\n\t\tof community 2{}\n\t\tand of the whole network:".format(op1,op2,op)
print "time to calculate slow function:{}\n".format(t1-t0) 
```

```python
# Global order parameter (fats method)
t2 = time.time()
op1 = global_order_parameter_spacetime(spacetime[:300,-300:]).mean()
op2 = global_order_parameter_spacetime(spacetime[300:,-300:]).mean()
op = global_order_parameter_spacetime(spacetime[:,-300:]).mean()
t3 = time.time()
print "order parameter of community 1:{}\n\t\tof community 2{}\n\t\tand of the whole network:".format(op1,op2,op)
print "time to calculate slow function:{}\n".format(t3-t2)
```

```python
# Local order parameter (one time snapshot at t=100)
lop = local_order_parameter_space(spacetime[:,100],4)
plt.plot(lop,'-')
plt.plot(lop,'o')
plt.show()
```
